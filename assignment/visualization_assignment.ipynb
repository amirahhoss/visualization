```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#1: Loading data
df = pd.read_csv('college_completion.csv',low_memory=False)

#2: Dimensions of data
print( df.shape, '\n')
#(3798,63) which means 3798 observations and 63 variables
print( df.head(), '\n')
#There are a lot of variables unlisted (...)

#3: Cross tabulate
print(pd.crosstab(df['control'],df['level']),'\n')
print(pd.crosstab(df['control'],df['level'], normalize = 'all'),'\n')
#Most of the colleges are 4 year private not-for-profit. The least of the colleges are 2 year private not-for-profit.
#Most of the institutions are public, or private-for-profit.

#4a: Histogram
import seaborn as sns
sns.histplot(df['grad_100_value'])
plt.show()

#4b: Kernel density plot
sns.kdeplot(df['grad_100_value'])
plt.show()

#4c: Boxplot
sns.boxplot(df['grad_100_value'])
plt.show()

#4d: Stat description
df['grad_100_value'].describe()

#5: Grouped kernel density plots
sns.kdeplot(data=df,x='grad_100_value',hue='control')
plt.show()

sns.kdeplot(data=df,x='grad_100_value',hue='level')
plt.show()

df.loc[:,['control','grad_100_value']].groupby('control').describe()
df.loc[:,['level','grad_100_value']].groupby('level').describe()

#4 year graduation rates are higher than 2 year.
#Private not-for-profit graduation rates are higher than public and private for-profit.

#6: New variable and kernel desnity plot

df['levelXcontrol']=df['level']+', '+df['control']
sns.kdeplot(data=df,x='grad_100_value',hue='levelXcontrol')
plt.show()
df.loc[:,['level','control','grad_100_value']].groupby(['level','control']).describe()

#Public 2 year graduation rates are the lowest
#Private for-profit 2 year graduation rates are 2x the rate for private for-profit 4 year

#7: Kernel density plot for aid_value
df['aid_arcsinh'] = np.arcsinh(df['aid_value'])
sns.kdeplot(data=df,x='aid_arcsinh')
plt.show()
sns.kdeplot(data=df,x='aid_value')
plt.show()
sns.kdeplot(data=df,x='aid_value',hue='control')
plt.show()
sns.kdeplot(data=df,x='aid_value',hue='level')
plt.show()

df.loc[:,['control','aid_value']].groupby('control').describe()
df.loc[:,['level','aid_value']].groupby('level').describe()

#private not-for-profit gives the most aid

#8: Scatter plots
sns.scatterplot(data=df,x='aid_value',y='grad_100_value')
plt.show()
sns.scatterplot(data=df,x='aid_value',y='grad_100_value',hue='control')
plt.show()
sns.scatterplot(data=df,x='aid_value',y='grad_100_value',hue='level')
plt.show()
sns.scatterplot(data=df,x='aid_value',y='grad_100_value',hue='levelXcontrol')
plt.show()

#Positive association is shown by 4 year public not-for-profit colleges in terms of aid and graduation rate
#lower aid = variability in graduation rates






```

```python
#3: Cleaning
df_subset = df[['RXS_GENERRX','RXS_DOSESRX','RXS_SKIPRX','RXS_NOFILLRX']]
print(df_subset.head())
```
